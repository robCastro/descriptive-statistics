```python
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')

iris.head()
```

Scatter plot

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length')
```

Doesn't seem like there's much of a tendency, however if we apply a hue we can see a pattern starting to emerge

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

Maybe the applied some kind of aggrupation algorithm? Perhaps some k nearest neigbhor

Joint Plot

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length')
```

This applies a histogram to the scatter plot, and we can apply a hue aswell so the data isn't as much confussing

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

Applying a hue also gives us the distribution curve instead of the histogram, and I can see that the sepal_lenght of the setosa species has a ver normal-like curve, however, its petal_lenght is very dense around where the mean probably is

Box plot

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
```

Virginica has an outlier
